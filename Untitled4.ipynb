```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datas = pd.read_csv('/content/IndiaPopulation_2021.csv')
datas.head()
```

```python
datas.shape
```

```python
X = datas[['Year','GrowthRate']]
Y = datas['Population']
```

```python
plt.plot(X,Y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
Xn,Xt,Yn,Yt = train_test_split(X,Y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(Xn,Yn)
```

```python
reg.score(Xt,Yt)
```

```python
pred = reg.predict(Xt)
reg.predict([[2040,0.6]])
```

```python
plt.scatter(Yt,pred)
```
